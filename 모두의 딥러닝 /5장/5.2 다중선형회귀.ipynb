```python
import numpy as np

x1 = np.array([2,4,6,8])
x2 = np.array([0,4,2,3])
y = np.array([81,93,91,97])
```

```python
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter3D(x1, x2, y)
plt.show()
```

```python
a1 = 0
a2 = 0
b = 0
lr = 0.01
epochs = 2001
n = len(x1)
```

```python
for i in range(epochs):
    y_pred = a1 * x1 +a2 * x2 + b 
    error = y - y_pred
    print(error)

    a1_diff = (2/n) * sum(-x1 * (error))
    a2_diff = (2/n) * sum(-x2 * (error))
    b_diff = (2/n) * sum(-(error))

    a1 = a1 - lr * a1_diff
    a2 = a2 - lr * a2_diff
    b = b - lr * b_diff

    if i % 100 == 0:
        print("epochs=%.f, 1=%.04f, 2=%.04f, =%.04f" %(i, a1, a2, b))

print(" : ", y)
print(" : ", y_pred)
```
